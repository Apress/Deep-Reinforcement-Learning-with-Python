```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu1 = -2
mu2 = 2
variance1 = 0.49
variance2 = 4
sigma1 = math.sqrt(variance1)
sigma2 = math.sqrt(variance2)
```

```python
x = np.linspace(mu1 - 3*sigma2, mu2 + 3*sigma2, 1000)
y1 = stats.norm.pdf(x, mu1, sigma1)
y2 = stats.norm.pdf(x, mu2, sigma1)
y3 = stats.norm.pdf(x, mu1, sigma2)
y4 = stats.norm.pdf(x, mu2, sigma2)


plt.figure(figsize=[16, 5])
plt.subplot(1,2,1)
plt.plot(x, y1)
plt.plot(x, y2)

plt.vlines(mu1, 0, np.max(y1), color='black', linestyle='--')
plt.vlines(mu2, 0, np.max(y2), color='black', linestyle='--')
plt.hlines(0.3, xmin=-2, xmax=2, color='red')

plt.subplot(1,2,2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.vlines(mu1, 0, np.max(y3), color='black', linestyle='--')
plt.vlines(mu2, 0, np.max(y4), color='black', linestyle='--')
plt.hlines(0.1, xmin=-2, xmax=2, color='red')

plt.show()
```
